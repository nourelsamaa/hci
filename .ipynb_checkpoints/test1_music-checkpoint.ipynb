```python
import mediapipe as mp # Import mediapipe
import cv2 # Import opencv
from dollarpy import Recognizer, Template, Point
import matplotlib.pyplot as plt

mp_drawing = mp.solutions.drawing_utils # Drawing helpers
mp_holistic = mp.solutions.holistic # Mediapipe Solutions
```

```python
templates=[]
```

```python
def getPoints(videoURL,label):
    cap = cv2.VideoCapture(videoURL)
    #web cam =0 , else enter filename
    # Initiate holistic model
    with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
        points = []
        left_index=[]
        left_thumb=[]
        right_index=[]
        right_thumb=[]
                
        while cap.isOpened():
            ret, frame = cap.read()

            # Recolor Feed
            if ret==True:

                image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                image.flags.writeable = False        

                # Make Detections
                results = holistic.process(image)
                # print(results.face_landmarks)


                # Recolor image back to BGR for rendering
                image.flags.writeable = True   
                image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

                # Drawing on Frame (You can remove it)
                # 2. Right hand
                mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                         mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                                         mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
                                         )

#                 # 3. Left Hand
                mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                          mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4),
                                          mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                                          )

                # 4. Pose Detections
                mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, 
                                         mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                                         mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                                         )
                # Export coordinates

                try:

                    # Does the frame include right hand ? 
                    try :
                        
                        if results.right_hand_landmarks != None :
                            right_index.append(Point(
                            results.right_hand_landmarks.landmark[8].x,
                            results.right_hand_landmarks.landmark[8].y,
                            1
                            ))
                            right_thumb.append(Point(
                            results.right_hand_landmarks.landmark[4].x,
                            results.right_hand_landmarks.landmark[4].y,
                            2
                            ))
                    except :
                        pass
                    # Does the frame include left hand ? 
                    try :

                        if results.left_hand_landmarks != None:
                            left_index.append(Point(
                            results.left_hand_landmarks.landmark[8].x,
                            results.left_hand_landmarks.landmark[8].y,
                            3
                            ))
                            left_thumb.append(Point(
                            results.left_hand_landmarks.landmark[4].x,
                            results.left_hand_landmarks.landmark[4].y,
                            4
                            ))
                    except : 
                        pass

                except:
                    print("error")
                    pass

                cv2.imshow(label, image)
#           Video ended (Doesn't have anymore frames)
            else :
                break
            if cv2.waitKey(10) & 0xFF == ord('q'):
                cap.release()
                cv2.destroyAllWindows()
                cv2.waitKey(100)
                break

    cap.release()
    cv2.destroyAllWindows()
    cv2.waitKey(100)
    
    points = right_thumb + right_index + left_index + left_thumb
    print(label)
    return points
```

```python
vid = "task2_dataset/indexN3.MP4"
points = getPoints(vid,"circle correct w/ index") 
points
```

```python
#Method 1 : All in one recognizer

#circle correct w/thumb n1
vid = "task2_dataset/thumbN1.MOV"
points = getPoints(vid,"circle correct w/thumb 1") 
tmpl_1 = Template('circle correct w/thumb 1', points)
templates.append(tmpl_1)

#circle correct w/thumb n2
vid = "task2_dataset/thumbN2.MP4"
points = getPoints(vid,"circle correct w/thumb 2") 
tmpl_2 = Template('circle correct w/thumb 2', points)
templates.append(tmpl_2)

#circle correct w/thumb n3
vid = "task2_dataset/thumbN3.MP4"
points = getPoints(vid,"circle correct w/thumb 3") 
tmpl_3 = Template('circle correct w/thumb 3', points)
templates.append(tmpl_3)

print("------------")

#circle correct w/index n1
vid = "task2_dataset/indexN1.MOV"
points = getPoints(vid,"circle correct w/index 1") 
tmpl_4 = Template('circle correct w/index 1', points)
templates.append(tmpl_4)

#circle correct w/index n2
vid = "task2_dataset/indexN2.MP4"
points = getPoints(vid,"circle correct w/index 2") 
tmpl_5 = Template('circle correct w/index 2', points)
templates.append(tmpl_5)

#circle correct w/index n3
vid = "task2_dataset/indexN3.MP4"
points = getPoints(vid,"circle correct w/index 3") 
tmpl_6 = Template('circle correct w/index 3', points)
templates.append(tmpl_6)
```

```python
#TEST
vid = "task2_dataset/R_indexN2.MP4"
points = getPoints(vid,"circle correct w/index 1") 



import time 
start = time.time()
recognizer = Recognizer(templates)
result = recognizer.recognize(points)
end = time.time()
print(result)
print(end-start)
```
